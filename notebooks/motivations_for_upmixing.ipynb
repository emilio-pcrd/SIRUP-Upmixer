# Motivations for Upmixing
This notebook shows why upmixing spatial audio represents an interesting challenge using SIRUP.

```python
import numpy as np
import pyroomacoustics as pra
from notebook_utils import RoomConfig, set_seed, make_room, add_microphones, simulate_room, compute_scm, compute_steering_vector, add_source
from plot_utils import plot_svect_magnitude, plot_waveform

set_seed(42)
```

## Setup Room & Simulate source
We initialize a room, add an array of microphones, and simulate the acoustic response when playing a signal.

```python
config = RoomConfig(room_dim=(6.0, 5.5, 4.5), rt60=0.3, snr_db=20.0)
room = make_room(config)
mic_positions = add_microphones(room, config)

# Create a dummy signal (noise in this case)
signal = np.random.randn(config.fs * 2)

# Add source and simulate
add_source(room, signal, (2.0, 3.0, 1.5))
signals = simulate_room(room, config.snr_db)

plot_waveform(signals[0, :], title='Received Signal (Mic 0)')
```

## Compute Steering Vector
We extract the steering vector from the received spatial audio.

```python
scm = compute_scm(signals)
svect = compute_steering_vector(scm)
plot_svect_magnitude(svect, title='Steering Vector Magnitude')
```
